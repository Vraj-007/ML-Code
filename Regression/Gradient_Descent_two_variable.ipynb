```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_regression(n_samples=1000, n_features=1, n_informative=1, noise=10, n_targets=1, random_state=10)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor

model = SGDRegressor()

model.fit(X, y)

print(model.coef_)
print(model.intercept_)

m = model.coef_
b = model.intercept_

plt.scatter(X, y)
plt.plot(X, m*X + b, color='red')
```

```python
class GDRegressor():

    def __init__(self, learning_rate, epochs):
        self.m = 100
        self.b = 10
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X, y):

        for i in range(self.epochs):

            b_slope = -2 * np.sum(y - self.m * X.ravel() - self.b)
            m_slope = -2 * np.sum((y - self.m * X.ravel() - self.b) * X.ravel())

            self.b = self.b - (self.lr * b_slope)
            self.m = self.m - (self.lr * m_slope)

            plt.plot(X, self.m*X + self.b)

            # print(f"{self.m}   {self.b}")

        plt.scatter(X, y)

        print(f"m = {self.m}")
        print(f"b = {self.b}")

    def coef_(self):
        return self.m

    def intercept_(self):
        return self.b
```

```python
model = GDRegressor(0.001, 50)

model.fit(X, y)
```

```python
m = model.coef_()
b = model.intercept_()

plt.scatter(X, y)
plt.plot(X, m*X + b, color='red')
```
