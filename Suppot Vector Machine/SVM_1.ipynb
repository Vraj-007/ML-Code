```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Use seaborn plotting defaults
import seaborn as sns ; sns.set() 
```

Working With Perfactly Linearly Separable Dataset

```python
# Generating dataset which is perfactly linearly separable

from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples = 200, centers=2, random_state=0, cluster_std=0.7)

plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='winter')
```

```python
from sklearn.svm import SVC     # Suppor Vector Classifier

model = SVC(kernel='linear', C=1)

model.fit(X, y)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)
print(f"R2 score : {r2_score(y, y_pred)}")
```

Working With Almost Linearly Separable Dataset

```python
# Generating dataset which is Almost linearly separable

from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples = 200, centers=2, random_state=0, cluster_std=1)

plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='winter')
```

```python
from sklearn.svm import SVC     # Suppor Vector Classifier

model = SVC(kernel='linear', C=10)

model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(f"R2 score : {r2_score(y, y_pred)}")
```
