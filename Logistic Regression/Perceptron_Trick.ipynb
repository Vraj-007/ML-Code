```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_classes=2, 
                            n_clusters_per_class=1, class_sep=10, hypercube=False, random_state=40)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=20)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def Perceptron(X,y):

    # Inserting 1 in first column of X for bias term
    X = np.insert(X, 0, 1, axis=1)
    # Initializing Weights
    weights = np.ones(X.shape[1])
    # Initializing Learning Rate
    lr = 0.01

    for i in range(1000):
        # Selecting random point
        j = np.random.randint(0,200)
        # Predicting output for selected random point
        y_hat = step(np.dot(X[j], weights))
        # Updating weights
        weights += (lr)*(y[j] - y_hat)*(X[j])

    return weights[0], weights[1:]
    # return bias, coef
```

```python
bias, coef = Perceptron(X, y)

print(f"Bias : {bias} \n")
print(f"Coef. : {coef} \n")
```

```python
m = -(coef[0]/coef[1])
b = -(bias/coef[1])
```

```python
x_input = np.linspace(-3, 3, 200) # 200 values between -3 and 3
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=30)
plt.ylim(-3, 3) # y limits
```
