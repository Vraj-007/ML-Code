```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples = 500, centers=4, random_state=42, cluster_std=1.5)

plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='winter')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=4, init="random")

model.fit(X_train)
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Centroid Initialization Method
good_init = np.array([[-7, -7], [-2, 10], [5, 2], [-9, 8]])

model = KMeans(n_clusters=4, init=good_init, n_init=1)

model.fit(X_train)
```

```python
# mean squared distance between each instance and its closest centroid

model.inertia_
```

```python
# centroid of each cluster 

model.cluster_centers_
```

```python
print(model.predict(model.cluster_centers_))
```

```python
print(X_test[0])

print(model.predict([X_test[0]]))
```

```python
# To know distance of datapoint from centroid of each cluster 

model.transform([X_test[0]])
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

accuracy_score(y_pred, y_test)
```

```python
count = 0

for i in range(len(X_test)):
    if(model.predict([X_test[i]]) != y_test[i]):
        count += 1
        print(f"y_pred : {model.predict([X_test[i]])}   &   y_test : {y_test[i]}")

print(f"{count}/{len(X_test)}")
```
