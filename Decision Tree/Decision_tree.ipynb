```python
import pandas as pd
from pandas_datareader import data
import numpy as np
from sklearn.tree import DecisionTreeRegressor # for regression problems
from sklearn.tree import DecisionTreeClassifier # for classification problem
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import load_iris
from sklearn.datasets import load_diabetes
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
iris = load_diabetes()
X = iris.data
y = iris.target
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 3], y)
```

```python
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)
```

```python
model = DecisionTreeRegressor(criterion='squared_error', max_leaf_nodes=20, min_samples_split=30
                                , splitter='random')


model.fit(X_train, y_train)
```

```python
model = DecisionTreeClassifier(criterion='gini', max_leaf_nodes=20, min_samples_split=30
                                , splitter='random')


model.fit(X_train, y_train)
```

```python
y_hat_1 = model.predict(X_train)
y_hat_2 = model.predict(X_test)

print(f"Training data r2 score : {r2_score(y_train, y_hat_1)}")
print(f"Testing data r2 score : {r2_score(y_test, y_hat_2)}")
```

```python
from sklearn.tree import plot_tree

# Plotting Decision Tree
plot_tree(model)
```
